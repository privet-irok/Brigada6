```python
# 1. Подключение нужных для работы моделей
import pandas as pd     # для чтения и записи табличных файлов
import numpy as np      # для математической обработки данных
import matplotlib.pyplot as plt # для построения графиков

from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook
```

```python
url = "https://drive.google.com/uc?export=download&id=1XFViS21mLmAnVUqzVsMYgxZUPw6sYn00"

input_df = pd.read_csv(url, header=0) 

input_df.head()
```

```python
input_df.tail()
```

```python
input_matrix = input_df.values # перевод таблицы в матрицу

print('размерность матрицы', input_matrix.shape)
print()

for i in range(0, len(input_df.columns)):
  print("{: <12} mean = {:.2f}".format(input_df.columns[i], input_matrix[: , i].mean()))
```

```python
sum1=0
for i in range(int(len(input_matrix)/5)):
  sum1+=input_matrix[i][0]
sum1=sum1/5
print(sum1)   
massive1=input_matrix[1752:3505]
sum2=0
for i in range(int(len(input_matrix)/5)):
  sum2+=massive1[i][0]
sum2=sum2/5
print(sum2)   
massive2=input_matrix[3504:5257]
sum3=0
for i in range(int(len(input_matrix)/5)):
  sum3+=massive2[i][0]
sum3=sum3/5
print(sum3)   
massive3=input_matrix[5256:7009]
sum4=0
for i in range(int(len(input_matrix)/5)):
  sum4+=massive3[i][0]
sum4=sum4/5
print(sum4)  
massive4=input_matrix[7008:8761]
sum5=0
for i in range(int(len(input_matrix)/5)):
  sum5+=massive4[i][0]
sum5=sum5/5
print(sum5)    
hope=[int(sum1), int(sum2), int(sum3), int(sum4), int(sum5)]
```

```python
plt.plot(hope)
plt.grid()
plt.ylabel('Мощность, Вт')
plt.xlabel("Время, с")
```

```python
output_notebook()
p = figure(title="Simple line example", x_axis_label='x', y_axis_label='y')
p.line(np.arange(len(hope)), hope, line_width = 2)
show(p)
```
